**DATA VISUALISATION**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

1. Frequency of each year.

```python
plt.figure()

jobs_in_data_df = pd.read_csv('/content/jobs_in_data.csv')

plt.figure()

sns.countplot(x = 'work_year', data = jobs_in_data_df, hue = 'work_year', palette='Spectral')
plt.title('Frequency of each year')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

My inference from the above chart:- The chart shows a gradual increase in data jobs every year between 2020 and 2023. There is a very significant increase in the frequency from 2022 to 2023.

2. Get Average Salaries for each year

```python
average_salary_per_year = jobs_in_data_df.groupby('work_year')['salary_in_usd'].mean()
plt.figure()
sns.barplot(x=average_salary_per_year.index, y=average_salary_per_year.values, hue = average_salary_per_year.index,
            palette='Spectral')
plt.title('Average Salaries for Each Year')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.show()
```

My inference from the above chart:- There is an upward trend in the average salaries over the years, with 2023 having the highest average salary. This increase might reflect the growing demand and valuation of data-related jobs over these years.

3. Get frequency of job titles

```python
plt.figure(figsize=(20, 5))
job_title_counts = jobs_in_data_df['job_title'].value_counts()
sns.barplot(y=job_title_counts.values, x=job_title_counts.index, hue=job_title_counts.values, palette='Spectral')
plt.title('Frequency of Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

My inference from the above chart:- The bar chart above visualizes the frequency of different data-related job titles in the dataset. The chart displays a variety of roles, with some being more common than others. The distribution of job titles is right-skewed, indicating that certain roles are more prevalent than others in the dataset.

4. Get the top 10 most common data jobs.

```python
top_10_job_titles = jobs_in_data_df['job_title'].value_counts().head(10)

plt.figure()
sns.barplot(x=top_10_job_titles.values, y=top_10_job_titles.index, hue=top_10_job_titles.index, palette='Spectral')
plt.title('Top 10 Most Common Data Job Titles')
plt.xlabel('Frequency')
plt.ylabel('Job Title')
plt.show()
```

My inference from the above chart:- Data Engineer and Data Scientist are the two most prevalent job titles in the data field, with Data Engineer being the most common. Data Analyst, Machine Learning Engineer, and Applied Scientist also appear frequently, indicating their significant presence in the data job market.

5. Visualize average salaries by job titles and the top job_titles with highest average salaries and least average salaries

```python
average_salaries = jobs_in_data_df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)

top_highest_salaries = average_salaries.head(5)
top_lowest_salaries = average_salaries.tail(5)
```

```python
plt.figure(figsize=(6, 25))
sns.barplot(x=average_salaries, y=average_salaries.index, hue=average_salaries.index)
plt.title('Average Salaries by Job Titles (in USD)')
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Titles')
plt.show()
```

```python
sns.barplot(x=top_highest_salaries, y=top_highest_salaries.index, hue=top_highest_salaries.index)
plt.title('Top 5 Highest Paying Job Titles')
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Titles')
plt.show()
```

```python
sns.barplot(x=top_lowest_salaries, y=top_lowest_salaries.index, hue=top_lowest_salaries.index)
plt.title('Top 5 Lowest Paying Job Titles')
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Titles')
plt.show()
```

6. Visualize Average Salaries for the most common job titles

```python
most_common_jobs = jobs_in_data_df['job_title'].value_counts().head(10).index

average_salaries_common_jobs = jobs_in_data_df[jobs_in_data_df['job_title'].isin(most_common_jobs)].groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)

plt.figure()
sns.barplot(x=average_salaries_common_jobs, y=average_salaries_common_jobs.index, hue=average_salaries_common_jobs.index)
plt.title('Average Salaries for the Most Common Job Titles (in USD)')
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Titles')
plt.show()
```

Insights from the data visualisation:- Some roles such as "Applied Scientist", "Machine Learning Engineer", and "Research Scientist" are among the highest-paying, with average salaries well above $180,000 USD. This reflects the high demand and specialised skills required in these fields. Moreover, positions like "Data Engineer" and "Data Analyst" show a notable decrease in average salary, indicating these might be more common entry to mid-level roles in the data field. Lastly, the chart demonstrates the diversity in job roles within the data industry, each commanding different levels of remuneration based on specialisation and demand.

(ii) Salary by Job Category and Experience Level:

The second plot illustrates the average salary across different job categories, segmented by experience level.  Some job categories, like Data Architecture and Modeling, show a significant increase in salary with experience. The plot also highlights the variance in average salaries across different job categories. Some categories, like Data Engineering and Data Science and Research, tend to have higher average salaries.
